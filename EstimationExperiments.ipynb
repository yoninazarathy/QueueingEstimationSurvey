```julia
using DataStructures, Distributions

"""
By default simulate an MD1 queue but if mm1 = true it is an MM1
"""
function simQueue(λ, m, T; Twarm = 0.0 , sampleTimes = [],  γ = 0.0, mm1 = false )
    tNextArr = rand(Exponential(1/(λ+γ)))
    tNextDep = Inf
    t = tNextArr
    qTotal = 0.0
    qRoomTotal = 0.0
    
    sampleIndex = 1
    samples = zeros(length(sampleTimes))
    tNextSam = length(sampleTimes) > 0 ? sampleTimes[sampleIndex] : Inf
    
    waitingRoom = Queue{Float64}()
    Q = 0
    waitTimes = Array{Float64,1}()
    probeSojournTimes = Array{Float64,1}()

    serT() = mm1 ? rand(Exponential(m)) : m
    
    while t < T + Twarm
        tPrev = t
        t = min(tNextArr,tNextDep,tNextSam)
        t > Twarm && (qTotal += (t - tPrev)*Q)
        Qr = Q ≥ 1 ? Q-1 : 0
        t > Twarm && (qRoomTotal += (t - tPrev)*Qr)

        if t == tNextArr
            if Q == 0
                st = serT()
                tNextDep = t + st
                Q = 1
                t > Twarm && push!(waitTimes, 0.0)
                t > Twarm && rand() ≤ γ/(γ+λ) && push!(probeSojournTimes, st)
            else
                enqueue!(waitingRoom,t)
                Q += 1
            end
            tNextArr = t + rand(Exponential(1/(λ+γ)))
        elseif t == tNextDep
            if length(waitingRoom) == 0
               tNextDep = Inf
               Q = 0
            else
                tArr = dequeue!(waitingRoom)
                Q -= 1
                waitTime = t - tArr
                t > Twarm && push!(waitTimes, waitTime)
                st = serT()
                t > Twarm && rand() ≤ γ/(γ+λ) && push!(probeSojournTimes, waitTime + st)
                tNextDep = t + st
            end
        else #sample 
            samples[sampleIndex] = Q
            sampleIndex += 1
            if sampleIndex > length(sampleTimes)
                tNextSam = Inf
            else
                tNextSam = sampleTimes[sampleIndex]
            end
        end
    end
    return  qTotal/T, 
            qRoomTotal/T, 
            mean(waitTimes), 
            mean(probeSojournTimes), 
            mean(samples),
            std(samples)
end
```

## Discrete Sampling Experiment - Confidence Intervals

```julia
using Distributions
quantile(Normal(),0.975)
```

```julia
function classicCI(xbar,sd,n)
    xbar - 1.96*sd/sqrt(n), xbar + 1.96*sd/sqrt(n)
end

function ciCoverage(Δ,n)
    λ, m = 0.9, 1.0
    Twarm = 10^3
    T = n*Δ
    output = simQueue(λ,m,T,Twarm = Twarm, sampleTimes = 0:Δ:T)
    ci = classicCI(output[5],output[6],n)
    ρ = λ*m
    md1TheorMean = (2-ρ)/2 * ρ/(1-ρ) 
    ci[1] <= md1TheorMean && md1TheorMean <= ci[2]    
end
```

```julia
using Random
M = 10^4
n = 100
ΔRange = 0.01:10:300.01#2:2:50#union(5:5:100, 100:20:200)
coverageVals =[]
for Δ in ΔRange
    Random.seed!(1)
    coverage = sum([ciCoverage(Δ,n) for _ in 1:M])/M
    push!(coverageVals,coverage)
    println((Δ,coverage))
end
println("done!")
```

```julia
using Plots
plot(ΔRange,coverageVals,
    xlabel="Sampling Interval Δ",
    ylabel="Probability",
    legend=:bottomright,
    label="Coverage Probability",
    lw = 2, c=:black, ylim = (0,1))
plot!([0,300],[0.95,0.95],c=:red,label="Desired 95% Confidence Level")
```

## Discrete Sampling Experiment - Estimator MSE

```julia
function mseSample(Δ,n)
    λ, m = 0.9, 1.0
    Twarm = 10^3
    T = n*Δ
    output = simQueue(λ,m,T,Twarm = Twarm, sampleTimes = 0:Δ:T)
    xBar = output[5] 
    λEst = 1+xBar -sqrt(1+xBar^2)
    (λEst - λ)^2   
end
```

```julia
using Random
M = 10^4
n = 100
ΔRange = 0.01:2.5:300.01#2:2:50#union(5:5:100, 100:20:200)
rmseEsts =[]
for Δ in ΔRange
    Random.seed!(1)
    rmse = sqrt(sum([mseSample(Δ,n) for _ in 1:M])/M)
    push!(rmseEsts,rmse)
    println((Δ,rmse))
end
println("done!")
```

```julia
using Random
λ = 0.90
m = 1.0
T = 10^3
Twarm = 10^
M = 10000

function probSampleMM1(γ)
    τ = simQueue(λ, m, T, γ = γ, Twarm = Twarm,mm1=true)[4]
    1-γ*m - m/τ
end

function probSampleMD1(γ)
    τ = simQueue(λ, m, T, γ = γ, Twarm = Twarm,mm1=false)[4]
    (τ - m)/(τ-m/2) - γ*m
end

mses = []
γRange = 0.002:0.002:0.098

for γ in γRange
    Random.seed!(0)
    @show γ
    ests = [probSampleMM1(γ) for _ in 1:M]
    mse = mean((ests .- λ).^2)
    push!(mses,mse)
end
println("done.")
```

```julia
using DataStructures,Distributions

function simulateMM1(lambda,mu,Q0,T)
    t, Q = 0.0 , Q0
    tValues = [0.0]
    qValues = [Q0]
    while t<T
        if Q == 0 #arrival to an empty system
            t += rand(Exponential(1/lambda))
            Q = 1
        else  #change of state when system is not empty 
            t += rand(Exponential(1/(lambda +mu)))
            Q += 2(rand() < lambda/(lambda+mu)) -1
        end
        push!(tValues,t)
        push!(qValues,Q)
    end
    return[tValues, qValues]
end

#This function returns a sequence of sojourn times of an M/M/1 queue
#that starts empty
function simulateMM1Sojourn(lambda,mu,T)
    tNextArr = rand(Exponential(1/lambda))
    tNextDep = Inf
    t = tNextArr
    
    waitingRoom = Queue(Float64)   
    arrTimeOfCustomerInService = NaN
    sojournTimes = []
    
    while t<T
        if t == tNextArr
            if isnan(arrTimeOfCustomerInService)
                tNextDep = t + rand(Exponential(1/mu))
                arrTimeOfCustomerInService = t
            else
                enqueue!(waitingRoom,t)
            end
            tNextArr = t + rand(Exponential(1/lambda))
        else #departure
            if length(waitingRoom) == 0
               tNextDep = Inf
               push!(sojournTimes, t-arrTimeOfCustomerInService)
               arrTimeOfCustomerInService = NaN
            else
               tNextDep = t + rand(Exponential(1/mu))
               push!(sojournTimes, t-arrTimeOfCustomerInService) 
               arrTimeOfCustomerInService = dequeue!(waitingRoom) 
            end
        end
        t = min(tNextArr,tNextDep)
    end
    
    return sojournTimes
end

function stichSteps(times,q)
    n = length(times)
    newTimes = [ times[1] ]
    newQ = [ q[1] ]
    for i in 2:n
        push!(newTimes,times[i])
        push!(newQ,q[i-1])
        push!(newTimes,times[i])
        push!(newQ,q[i])
    end
    return [newTimes,newQ]
end

function sampleQ(times,qValues,delta)
    t = 0.0 #time of sampling
    tMax = last(times)
    samples = []
    i = 2 #index in times
    while t<tMax
        while times[i]<t
           i += 1 
        end
        push!(samples,qValues[i-1])
        t += delta
    end
    return samples
end

function findArrDep(times,qValues)
    n = length(times)
    arrTimes = []
    depTimes = []
    for i in 2:n
        if qValues[i] > qValues[i-1]
            push!(arrTimes,times[i])
        else
            push!(depTimes,times[i])
        end
    end
    return [arrTimes,depTimes]
end
```
